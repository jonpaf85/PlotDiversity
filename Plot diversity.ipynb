```python
#BAR PLOT
# make the stats
sample_mean = df.mean()
sample_std = df.std()

# Build the Bar plot
fig, ax = plt.subplots()

ax.bar(df.columns,
       sample_mean,
       yerr=sample_std,
       color=['green', 'purple'],
       align='center',
       ecolor='black',
       capsize=10)
ax.set_ylabel('Values')
```
